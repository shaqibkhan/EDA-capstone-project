# **5.handling null values**

````python
data1=data1.drop('company',axis=1)
````

````python
data1.duplicated().sum()
````

````python
data1.describe()
````

#**7.handling outliers**

````python
sns.boxplot(y="adr",data=data1)
plt.title('box plot for finding outliers in adr column')
plt.show()
````

````python
q1=data1['adr'].quantile(0.25)
q3=data1['adr'].quantile(0.75)
````

````python
IQR=q3-q1
IQR
````

````python
l=q1-(1.5*IQR)
l
````

````python
h=q3+(1.5*IQR)
h
````

````python
data=data1[(data1['adr']>=l) & (data1['adr']<=h)]
````

````python
sns.boxplot(y='adr',data=data)
````
